```python
import mysql.connector
import pandas as pd

mydb = mysql.connector.connect(
  host="localhost",
  user="ragu",
  password="password"
)

mycursor = mydb.cursor()
```

```python
# Create Database 
mycursor.execute("create database joins")
```

```python
# SHOW DATABASES
mycursor.execute("SHOW DATABASES")

for x in mycursor:
    print(x) 
```

```python
# Use Joins db
mycursor.execute("use joins")
```

```python
# Load Datasets to DB-Iris
data = pd.read_csv("../Datasets/Iris.csv")
data
```

```python
# Create Table Iris-sepal
mycursor.execute("create Table Iris_sepal(Id INT PRIMARY KEY, SepalLengthCm FLOAT , SepalWidthCm FLOAT )")
```

```python
# Insert into Iris_sepal
for row in data.itertuples():
    sql = "INSERT INTO Iris_sepal (Id, SepalLengthCm, SepalWidthCm) VALUES (%(Id)s, %(SepalLengthCm)s, %(SepalWidthCm)s)"
    val = {
        "Id":row.Id,
        "SepalLengthCm":row.SepalLengthCm,
        "SepalWidthCm":row.SepalWidthCm    
    }  
    mycursor.execute(sql, val)
    mydb.commit()
```

```python
# Create Table Iris-petal
mycursor.execute("create Table Iris_petal(Id INT PRIMARY KEY,  PetalLengthCm FLOAT, PetalWidthCm FLOAT )")
```

```python
# Insert into Iris_petal
for row in data.itertuples():
    sql = "INSERT INTO Iris_petal (Id,  PetalLengthCm, PetalWidthCm ) VALUES (%(Id)s, %(PetalLengthCm)s, %(PetalWidthCm)s)"
    val = {
        "Id":row.Id,
        "PetalLengthCm":row.PetalLengthCm,
        "PetalWidthCm": row.PetalWidthCm   
    }  
    mycursor.execute(sql, val)
    mydb.commit()
```

```python
# Create Table Iris_species
mycursor.execute("create Table Iris_species(Id INT PRIMARY KEY, Species VARCHAR(16))")
```

```python
# Insert into Iris_species
for row in data.itertuples():
    sql = "INSERT INTO Iris_species (Id,Species) VALUES (%(Id)s,%(Species)s)"
    val = {
        "Id":row.Id,
        "Species":row.Species     
    }  
    mycursor.execute(sql, val)
    mydb.commit()
```
