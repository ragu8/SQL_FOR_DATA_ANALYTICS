```python
import mysql.connector
import pandas as pd

# Database Connection
mydb = mysql.connector.connect(host = "localhost",user = "ragu",password ="password")
cursor = mydb.cursor()
```

```python
# Use Database joins
cursor.execute("use joins")
```

```python
# List of Tables
cursor.execute("show tables")
for x in cursor:
    print(x)
```

```python
# Retrieving data using select
pd.read_sql_query("select * from Iris_petal ",mydb)
```

```python
# Retrieving data using select
pd.read_sql_query("select * from Iris_sepal ",mydb)
```

```python
# Retrieving data using select
pd.read_sql_query("select * from Iris_species ",mydb)
```

```python
# Inner join
pd.read_sql_query("select PetalLengthCm , PetalWidthCm, Species from Iris_petal INNER JOIN Iris_species on Iris_petal.Species_Id = Iris_species.Species_Id ",mydb) 
```

```python
pd.read_sql_query("select * from Iris_petal INNER JOIN Iris_species on Iris_petal.Species_Id = Iris_species.Species_Id order by 1",mydb) 
```

```python
pd.read_sql_query("select * from Iris_petal INNER JOIN Iris_species on Iris_petal.Species_Id = Iris_species.Species_Id where Iris_species.species ='Iris-setosa'  ",mydb) 
```

```python
pd.read_sql_query("select * from Iris_petal INNER JOIN Iris_species on Iris_petal.Species_Id = Iris_species.Species_Id  where Iris_species.species ='Iris-setosa' order by 3  ",mydb) 
```

```python
pd.read_sql_query("select * from Iris_petal INNER JOIN Iris_species on Iris_petal.Species_Id = Iris_species.Species_Id  where Iris_species.species ='Iris-setosa' order by 3 limit 10 ",mydb) 
```

```python
pd.read_sql_query("select * from Iris_petal P INNER JOIN Iris_species S on P.Species_Id = S.Species_Id  where S.species ='Iris-setosa' order by 3 limit 5  ",mydb) 
```

```python
# Outer join
# Left outer join
pd.read_sql_query("select PetalWidthCm , sepalWidthCm from Iris_petal P  LEFT OUTER JOIN  Iris_sepal S on P.Id = S.Id  limit  10 ",mydb)
```
